```python
# Check for missing values in each column
missing_values = Covid_19_Indonesia_Dataset_df.isnull().sum()

# Percentage of missing values
missing_percentage = (missing_values / len(Covid_19_Indonesia_Dataset_df)) * 100

missing_summary = pd.DataFrame({'Missing Values': missing_values, 'Percentage': missing_percentage})
missing_summary[missing_summary['Missing Values'] > 0]
```

```python
# Impute missing values in 'Province' and 'Time Zone' with "Not Applicable"
Covid_19_Indonesia_Dataset_df['Province'].fillna('Not Applicable', inplace=True)
Covid_19_Indonesia_Dataset_df['Time Zone'].fillna('Not Applicable', inplace=True)

# Verify the imputation
missing_values_after = Covid_19_Indonesia_Dataset_df.isnull().sum()

missing_values_after
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Convert the Date column to datetime format
Covid_19_Indonesia_Dataset_df['Date'] = pd.to_datetime(Covid_19_Indonesia_Dataset_df['Date'])

# Filter the data for country-level records
country_data = Covid_19_Indonesia_Dataset_df[Covid_19_Indonesia_Dataset_df['Location Level'] == 'Country']

# Aggregate data by date for the whole country
country_agg = country_data.groupby('Date').sum()

# Plotting
plt.figure(figsize=(14, 8))

# Total Cases
plt.plot(country_agg.index, country_agg['Total Cases'], label='Total Cases', color='blue')

# Total Deaths
plt.plot(country_agg.index, country_agg['Total Deaths'], label='Total Deaths', color='red')

# Total Recovered
plt.plot(country_agg.index, country_agg['Total Recovered'], label='Total Recovered', color='green')

plt.title('COVID-19 Trends in Indonesia (Country Level)')
plt.xlabel('Date')
plt.ylabel('Number of Cases')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

```python
# Filter data for provincial level records
province_data = Covid_19_Indonesia_Dataset_df[Covid_19_Indonesia_Dataset_df['Location Level'] == 'Province']

# Latest data per province
latest_data = province_data.sort_values('Date').groupby('Province').last()

# Top 5 provinces with the highest number of cases, deaths, and recoveries
top_cases = latest_data.nlargest(5, 'Total Cases')
top_deaths = latest_data.nlargest(5, 'Total Deaths')
top_recoveries = latest_data.nlargest(5, 'Total Recovered')

# Plotting
fig, axs = plt.subplots(3, 1, figsize=(10, 15))

# Top Cases
sns.barplot(x=top_cases['Total Cases'], y=top_cases.index, ax=axs[0], palette="Blues_d")
axs[0].set_title('Top 5 Provinces by Total Cases')

# Top Deaths
sns.barplot(x=top_deaths['Total Deaths'], y=top_deaths.index, ax=axs[1], palette="Reds_d")
axs[1].set_title('Top 5 Provinces by Total Deaths')

# Top Recoveries
sns.barplot(x=top_recoveries['Total Recovered'], y=top_recoveries.index, ax=axs[2], palette="Greens_d")
axs[2].set_title('Top 5 Provinces by Total Recoveries')

plt.tight_layout()
plt.show()
```
